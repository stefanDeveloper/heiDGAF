```python
import pandas as pd
```

```python
df_heicloud = pd.read_csv("data/heicloud/2023-10-15_2023-10-16_sorted_heiCLOUD_DNS_responses.txt", sep=" ")
df_heicloud
```

```python
from dgad.prediction import Detective


mydomains = df_heicloud.to_numpy().reshape(-1).tolist()

print(mydomains)

detective = Detective()
# convert mydomains strings into dgad.schema.Domain
mydomains, _ = detective.prepare_domains(mydomains)
# classify them
detective.investigate(mydomains)
```

```python
with open('./result.json', 'w') as f_new:
    f_new.write(str(mydomains))
```
